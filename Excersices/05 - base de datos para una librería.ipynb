# Creación de una base de datos para una librería

## Escenario:
Miguel Ángel García es propietario de una pequeña librería llamada "El Rincón de los Libros". Desea modernizar su sistema de gestión de clientes mediante la creación de una base de datos digital. En este ejercicio, aprenderemos cómo crear una base de datos, una tabla y cómo insertar datos en ella para ayudar a Miguel Ángel con su objetivo.

## Paso a paso:

1. **Crear la base de datos:**
   - En el gestor de bases de datos, accede al sistema utilizando tus credenciales.
   - Utiliza el siguiente comando SQL para crear la base de datos llamada "libreria":
     ```sql
     CREATE DATABASE libreria;
     ```

2. **Seleccionar la base de datos:**
   - Una vez creada la base de datos, selecciona la base de datos recién creada utilizando el comando SQL:
     ```sql
     USE libreria;
     ```

3. **Crear la tabla de clientes:**
   - Ahora, vamos a crear la tabla "clientes" que almacenará la información de los clientes. Utiliza el siguiente comando SQL:
     ```sql
     CREATE TABLE clientes (
         id_cliente INT,
         nombre_cliente VARCHAR(50),
         direccion_cliente VARCHAR(255)
     );
     ```

4. **Insertar datos en la tabla:**
   - A continuación, insertaremos los datos del primer cliente en la tabla. Utiliza este comando SQL:
     ```sql
     INSERT INTO clientes (id_cliente, nombre_cliente, direccion_cliente) VALUES (1, 'Juan', 'Calle Principal 123, Ciudad');
     ```

5. **Verificar la inserción de datos:**
   - Para asegurarnos de que los datos se han insertado correctamente, podemos ejecutar una consulta para ver el contenido de la tabla. Utiliza el siguiente comando SQL:
     ```sql
     SELECT * FROM clientes;
     ```

6. **Insertar otro registro de cliente:**
   - Como tarea adicional, vamos a insertar los datos de otro cliente en la tabla. Utiliza este comando SQL:
     ```sql
     INSERT INTO clientes (id_cliente, nombre_cliente, direccion_cliente) VALUES (2, 'María', 'Avenida Central 456, Ciudad');
     ```

7. **Verificar la inserción del nuevo registro:**
   - Para confirmar que el nuevo registro se ha añadido correctamente, ejecuta nuevamente la consulta:
     ```sql
     SELECT * FROM clientes;
     ```

Con estos pasos, has creado una base de datos para la librería "El Rincón de los Libros", has creado una tabla para almacenar información de los clientes y has insertado datos en ella.
