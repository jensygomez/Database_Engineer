```python
# Importa el módulo pymysql para trabajar con bases de datos MySQL
import pymysql
import random

# Lista de nombres y apellidos predefinidas
nombres_brasilenos = [
    'Antonio', 'João', 'Francisco', 'José', 'Luiz', 'Carlos', 'Paulo', 'Pedro',
    'Maria', 'Ana', 'Francisca', 'Antonia', 'Marcia', 'Adriana', 'Juliana', 'Luciana',
    'Rafael', 'Daniel', 'Fernando', 'Marcelo'
]

apellidos_brasilenos = [
    'Silva', 'Santos', 'Oliveira', 'Souza', 'Pereira', 'Costa', 'Carvalho', 'Almeida',
    'Ferreira', 'Pinto', 'Gonçalves', 'Rodrigues', 'Lima', 'Alves', 'Gomes', 'Martins',
    'Araújo', 'Rocha', 'Ribeiro', 'Barbosa'
]

# Lista de personajes importantes en la historia brasileña
personajes_brasilenos = [
    'Pedro Álvares Cabral', 'Tiradentes', 'Getúlio Vargas', 'Emperor Pedro II', 'Princess Isabel',
    'Oscar Niemeyer', 'Juscelino Kubitschek', 'Machado de Assis', 'Heitor Villa-Lobos', 'Carmen Miranda',
    'Chico Mendes', 'Ayrton Senna', 'Pelé', 'Oswaldo Cruz', 'Zumbi dos Palmares', 'Ruy Barbosa', 'Tom Jobim',
    'Cândido Rondon', 'Carlos Drummond de Andrade', 'Baron of Rio Branco'
]

# Funciones para generar nombres, apellidos y direcciones aleatorias
def generar_nombres_aleatorios(cantidad):
    return random.choices(nombres_brasilenos, k=cantidad)

def generar_apellidos_aleatorios(cantidad):
    return random.choices(apellidos_brasilenos, k=cantidad)

def generar_direcciones_aleatorias(cantidad):
    return random.choices(personajes_brasilenos, k=cantidad)

# Función para conectar a la base de datos MySQL
def conectar_base_datos():
    try:
        # Establece una conexión a la base de datos MySQL        
        connection = pymysql.connect(
            host='sql10.freesqldatabase.com',
            user='sql10705361',
            password='HqKECAGFRn',
            database='sql10705361'
        )
        print("Conexión establecida correctamente")
        return connection
    except pymysql.Error as e:
        print("Error al intentar conectarse a la Base de Datos:", e)
        return None

# Función para crear la tabla en MySQL
def crear_tabla(connection):
    try:
        cursor = connection.cursor()
        consulta = """
        CREATE TABLE IF NOT EXISTS nombre_tabla (
            id INT AUTO_INCREMENT PRIMARY KEY,
            nombre TEXT,
            apellido TEXT,
            direccion TEXT
        )
        """
        cursor.execute(consulta)
        connection.commit()
        print("Tabla creada correctamente")
        return cursor
    except pymysql.Error as e:
        print("Error al crear la tabla:", e)
        return None

# Función para llenar la base de datos
def llenar_base_de_datos(connection, cantidad):
    try:
        cursor = connection.cursor()
        nombres = generar_nombres_aleatorios(cantidad)
        apellidos = generar_apellidos_aleatorios(cantidad)
        direcciones = generar_direcciones_aleatorias(cantidad)

        for nombre, apellido, direccion in zip(nombres, apellidos, direcciones):
            cursor.execute('INSERT INTO nombre_tabla (nombre, apellido, direccion) VALUES (%s, %s, %s)', (nombre, apellido, direccion))
        
        connection.commit()
        print("Valores insertados correctamente.")
    except pymysql.Error as e:
        print("Error al insertar valores en la tabla:", e)

# Crear conexión a la base de datos MySQL
connection = conectar_base_datos()

# Llamada a la función para crear la tabla en MySQL
crear_tabla(connection)

# Llamada a la función para llenar la tabla en MySQL
cantidad = 10000  # Cantidad de registros a generar
llenar_base_de_datos(connection, cantidad)

# Cerrar conexión a la base de datos MySQL
connection.close()
```
