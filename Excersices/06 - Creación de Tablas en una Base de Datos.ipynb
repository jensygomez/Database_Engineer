### Objetivo:
En este ejercicio, aprenderás a crear una tabla en una base de datos desde cero. El objetivo es construir una tabla en una base de datos para almacenar los datos personales básicos de los jugadores de fútbol menores de 16 años.

### Escenario:

### Instrucciones:

#### Paso 1: Crear una base de datos
1. Crea una nueva base de datos llamada `football_club` con el siguiente comando SQL:
   ```sql
   CREATE DATABASE football_club;
   ```

2. Selecciona la base de datos recién creada con el siguiente comando:
   ```sql
   USE football_club;
   ```

#### Paso 2: Crear una tabla
1. Escribe una sentencia SQL que contenga el comando `CREATE TABLE` seguido del nombre de la tabla. En este caso, `Players` es un nombre adecuado.
2. Abre paréntesis para definir las columnas de la tabla:
   - `playerID`
   - `playerName`
   - `age`
3. Asigna a cada columna un tipo de datos adecuado:
   - `playerID`: INT
   - `playerName`: VARCHAR(50)
   - `age`: INT
4. Una vez que todas las columnas requeridas estén definidas, agrega un paréntesis de cierre y un punto y coma al final de la sentencia SQL de la siguiente manera:
   ```sql
   CREATE TABLE players (playerID INT, playerName VARCHAR(50), age INT);
   ```
